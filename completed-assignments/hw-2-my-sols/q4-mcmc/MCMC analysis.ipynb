* The posterior histogram of the latent variable $home$.
* The scatter plot containing the estimates of the posterior mean attacking strength $\mathbb{E}_{p(\boldsymbol{\theta}, \boldsymbol{\eta} | \mathbf{y})}[att_i]$ against the estimates of the posterior mean defending strength $\mathbb{E}_{p(\boldsymbol{\theta}, \boldsymbol{\eta} | \mathbf{y})}[def_i]$ for 20 teams where $i = 0, ..., 19$. 

```python
# Import modules.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import os
import sys
```

```python
# Functions used to generate the required visualisations.

def create_dict():
    """Construct team_idx -> team name dictionary

    Returns
    -------
    team_idx_name_dict : Dict[int, str]
        A dictionary of (team_ix -> team name) key-value pairs.
    """

    team_idx = [idx for idx in range(20)]
    names = ['Arsenal', 'Aston Villa', 'Cardiff City', 'Chelsea', 'Crystal Palace',
             'Everton', 'Fulham', 'Hull City', 'Liverpool', 'Manchester City',
             'Manchester United', 'Newcastle United', 'Norwich City', 'Southampton', 'Stoke City',
             'Sunderland', 'Swansea City', 'Tottenham Hotspurs', 'West Bromwich Albion', 'West Ham United']

    team_idx_name_dict = dict(zip(team_idx, names))

    return team_idx_name_dict

def generate_histogram(png_filepath, posterior_samples):
    
    plt.style.use('seaborn')
    
    fig, ax = plt.subplots(figsize=(10,8))

    num_samples = posterior_samples.shape[0]

    # Use Freedman-Diaconis rule to compute histogram bin-width.
    q25, q75 = np.percentile(posterior_samples, [25, 75])
    FD_bin_width = (2 * (q75 - q25)) / (num_samples ** (1 / 3))
    num_bins = round((posterior_samples.max() - posterior_samples.min()) / FD_bin_width).astype(int)
    
    # Plot histogram.
    ax.hist(posterior_samples, density=True, bins=num_bins)
    ax.set_xticks(np.linspace(-0.04, 0.06, num=11))

    # Overlay a kernel density estimate of the PDF using a Gaussian kernel
    # with an automatically configured bandwidth for illustrative purposes only.
    # Use posterior samples as dataset to instantiate Gaussian KDE object,
    # then evaluate it over a grid of points.
    x_min, x_max = ax.get_xlim()
    x = np.linspace(x_min, x_max, 3000)
    kde = gaussian_kde(posterior_samples, bw_method='silverman')

    ax.plot(x, kde.evaluate(x), label="Gaussian KDE")

    # Compute and indicate the empirical mean of the posterior.
    y_min, y_max = ax.get_ylim()
    emp_mean = np.mean(posterior_samples)
    ax.vlines(emp_mean,
              ymin=y_min, ymax=kde.evaluate(emp_mean),
              colors='k', linewidth=0.8, linestyle='dotted',
              label=r'Est. posterior mean $\approx$ {}'.format(np.round(emp_mean, 3)))

    # Axes guidelines.
    ax.axvline(0, color='black', linewidth=0.8)
    ax.axhline(0, color='black', linewidth=1.5)

    # Plot info.
    ax.set_title(r"Posterior density histogram of latent variable $home$, $N = 5000$ samples.",
                 fontsize=13)
    ax.set_xlabel(r"$home$")
    ax.set_ylabel("Probability density")
    ax.legend(loc='upper right')
    
    # Save figure.
    fig.savefig(png_filepath)
    
    
def generate_scatterplot(png_filepath, posterior_samples, team_idx_name_dict):
    
    plt.style.use('seaborn')
    
    fig, ax = plt.subplots(figsize=(16, 9))
    
    # Compute the empirical posterior means of each team's attacking and
    # defending strength.
    emp_mean_atts = np.mean(posterior_samples[:, 1:21], axis=0)
    emp_mean_defs = np.mean(posterior_samples[:, 21:41], axis=0)

    # Plot each team's estimated mean attack and defence on a scatter plot.
    for team_idx in range(20):
        team = team_idx_name_dict[team_idx]
        label = str(team_idx) + ". " + str(team)
        ax.scatter(emp_mean_defs[team_idx], emp_mean_atts[team_idx],
                   label=label)

    # Colours hard to see - annotate each point with team idx.
    for team_idx in range(20):
        ax.annotate(team_idx,
                    (emp_mean_defs[team_idx] + 0.0005, emp_mean_atts[team_idx] + 0.0005))

    # Axes guidelines
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

    # Annotate scatterplot.
    ax.set_title('Scatter plot: '
                 + 'Estimated posterior mean attacking strengths $\widebar{att}_i$ '
                 + 'against estimated posterior mean defending strengths $\widebar{def}_i$ '
                 + "\n"
                 + 'for 20 Premier League teams.',
                 fontsize=14)

    ax.set_xlabel('Estimated posterior mean defending strength ' + r'$\widebar{def_i}$', fontsize=12)
    ax.set_ylabel('Estimated posterior mean attacking strength ' + r'$\widebar{att_i}$', fontsize=12)

    ax.legend(loc='upper left')
    
    # Save figure.
    fig.savefig(png_filepath)
```

```python
# Load the .npz file from the optimal parameter setting.
# This was ascertained from "HW2-10-708.ipynb" to be sigma = 0.005, t = 50.
# Assuming the 'metropolis.py' script has been run correctly, the .npz file
# should be stored in a 'results' folder in current root directory.

npz_filename = 'metropolis-results-sigma=0.005-t=50.npz'

path = os.getcwd()
results_dir_name = "results"
results_dir_path =  os.path.join(path, results_dir_name)
npzfile_path = os.path.join(results_dir_path, npz_filename)
npz_file = np.load(npzfile_path, allow_pickle=True)
npz_file.files
```

```python
# Generate the posterior density histogram for the latent variable home.
# Save the figure to '/results' directory.

png_filename = 'posterior-density-histogram-home.png'
png_filepath = os.path.join(results_dir_path, png_filename)

posterior_samples_home = npz_file['theta_posterior_samples'][:, 0]

generate_histogram(png_filepath, posterior_samples_home)
```

```python
# Generate the scatter plot and save to '/results'.

png_filename = 'scatter-plot-estimated-posterior-means.png'
png_filepath = os.path.join(results_dir_path, png_filename)

team_idx_name_dict = create_dict()
posterior_samples = npz_file['theta_posterior_samples']

generate_scatterplot(png_filepath, posterior_samples, team_idx_name_dict)
```
